```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

%matplotlib inline
```

```python
df = pd.read_csv("amz_uk_price_prediction_dataset.csv")
```

```python
print(df.head())
print(df.dtypes)
print(df.shape)
```

```python
print(df['stars'].value_counts())
print(df['isBestSeller'].value_counts())

frequency_table = df['category'].value_counts().head(5)
```

```python
# top 5 most listed product categories
print(df['category'].value_counts().head(5))
```

```python
ax=sns.barplot(y=frequency_table.index, x=frequency_table.values, palette="Set3", hue=frequency_table.index, legend='full')
sns.move_legend(ax, "lower right")
```

```python
df['category'].value_counts().head(3).plot.pie(autopct='%1.1f%%', startangle=30, colors=sns.color_palette("Set3"));
```

```python
print(df['price'].mean())
print(df['price'].median())
print(df['price'].mode())
```

```python
# average price if a lot higher than most common price, meaning you have outliers
```

```python
df.describe()
```

```python
print(df['price'].std())
print(df['price'].var())
print(df['price'].quantile([0.25,0.75]))
```

```python
sns.histplot(df['price'], kde=True, bins=100, color="salmon");

```

```python
sns.boxplot(x = df['price'], color="lightblue");
```

```python
df['price_quantile'] = pd.qcut(df['price'], q=4, labels=['Q1', 'Q2', 'Q3', 'Q4'])

df.price_quantile.value_counts()
```

```python
sns.histplot(df.price_quantile.value_counts(), kde=True, bins=30, color="salmon");
```

```python
print(df['stars'].mean())
print(df['stars'].median())
print(df['stars'].mode())
print(df['stars'].std())
print(df['stars'].var())
print(df['stars'].quantile([0.25,0.75]))
```

```python
# ratings are highly variable, towards low end
```
